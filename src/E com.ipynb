```python
#RFM Analysis (Recency, Frequency, Monetary)
```

```python
max_date_str = df1['Purchase Date'].max()

customers_last_purchase_dates = df1.groupby('Customer Name')['Purchase Date'].max().reset_index()

max_date = pd.to_datetime(max_date_str)

customers_last_purchase_dates = df1.groupby('Customer Name')['Purchase Date'].max().reset_index()

customers_last_purchase_dates['Purchase Date'] = pd.to_datetime(customers_last_purchase_dates['Purchase Date'])

customers_last_purchase_dates['Recency'] = (max_date - customers_last_purchase_dates['Purchase Date']).dt.days
```

```python
print(customers_last_purchase_dates.head())
```

```python
df1 = pd.merge(df1, customers_last_purchase_dates[['Customer Name', 'Recency']], on = 'Customer Name', how = 'left')

df1
```

```python
#Recency

recency_df = df1.groupby('Customer Name')['Recency'].min().reset_index()

recency_df
```

```python
#Customers Frequency

frequency_df = df1.groupby('Customer Name')['Purchase Date'].count().reset_index()

frequency_df.rename(columns = {'Purchase Date': 'Frequency'}, inplace = True)

frequency_df
```

```python
#Customers Monetary Value (total amount spent by each customer)

monetary_df = df1.groupby('Customer Name')['Total Purchase Amount'].sum().reset_index()

monetary_df.rename(columns={'Total Purchase Amount': 'Monetary'}, inplace=True)

monetary_df
```

```python
rfm_df = pd.merge(recency_df, frequency_df, on = 'Customer Name')

rfm_df
```

```python
rfm_df = pd.merge(rfm_df, monetary_df, on='Customer Name')

rfm_df
```

```python
#Churn Analysis
```

```python
df1['Total Spending'] = df1['Product Price'] * df1['Quantity']

df1['Total Spending']
```

```python
df1['First Purchase Date'] = df1.groupby('Customer Name')['Purchase Date'].transform('min')

df1['First Purchase Date']
```

```python
df1['Purchase Date'] = pd.to_datetime(df1['Purchase Date'])

df1['First Purchase Date'] = pd.to_datetime(df1['First Purchase Date'])
```

```python
df1['Customer Tenure'] = (df1['Purchase Date'].max() - df1['First Purchase Date']).dt.days

df1['Customer Tenure']
```

```python
customer_df = df1.groupby('Customer Name').agg({
    'Total Spending': 'sum',
    'Customer Tenure': 'max',
    'Returns': 'sum',
    'Churn': 'max'
}).reset_index()

customer_df
```

```python
churn_rate = customer_df['Churn'].value_counts(normalize = True) * 100

print(f"Churn Rate: {churn_rate.iloc[1]:.2f}%")

#or

#churn_rate_churned = churn_rate.get(1, 0)

#print(f"Churn Rate: {churn_rate_churned:.2f}%")
```
